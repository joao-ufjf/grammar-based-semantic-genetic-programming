```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

from __future__ import division
import random
from sympy import *

IF_ELSE = set(['if-else'])
BINARY_OPERATORS = set(['+', '-', '*', '/', '^', '<', '<=', '==', '>=', '>', '!=', 'and', 'or'])
UNARY_OPERATORS = set(['log', 'sqrt', 'exp', 'sin'])

### POSTFIX ===> INFIX ###
'''
1) When see an operand, push
2) When see an operator, pop out two numbers, connect them into a substring and push back to the stack
3) the top of the stack is the final infix expression.
'''
def postfix_to_infix(formula):
    stack = []
    for ch in formula.split():
        if ch == "<-":
            b = stack.pop()
            a = stack.pop()
            expr = "v[" + a + "] = " + b + ""
            stack.append(expr)
        elif ch == "->":
            a = stack.pop()
            expr = "v[" + a + "]"
            stack.append(expr)
        elif ch in BINARY_OPERATORS:
            if ch == "^":
                ch = "**"
            b = stack.pop()
            a = stack.pop()
            expr = "(" + a + ch + b + ")"
            stack.append(expr)
        elif ch in UNARY_OPERATORS:
            a = stack.pop()
            expr = ch + "(" + a + ")"
            stack.append(expr)
        elif ch in IF_ELSE:
            c = stack.pop()
            b = stack.pop()
            a = stack.pop()
            expr = "if(" + c + "){\n" + a + "\n}else{\n" + b + "\n}"
            stack.append(expr)
        else:
            stack.append(ch)

    return stack

def postfix_to_infix_python(formula, constants):
    stack = []
    c_counter = 0
    for ch in formula.split():
        if ch == "<-":
            b = stack.pop()
            a = stack.pop()
            expr = "v[" + a + "] = " + b + ""
            stack.append(expr)
        elif ch == "->":
            a = stack.pop()
            expr = "v[" + a + "]"
            stack.append(expr)
        elif ch in BINARY_OPERATORS:
            if ch == "^":
                ch = "**"
            b = stack.pop()
            a = stack.pop()
            expr = "(" + a + ch + b + ")"
            stack.append(expr)
        elif ch in UNARY_OPERATORS:
            a = stack.pop()
            expr = ch + "(" + a + ")"
            stack.append(expr)
        elif ch in IF_ELSE:
            c = stack.pop()
            b = stack.pop()
            a = stack.pop()
            expr = "if(" + c + "):\n" + a + "\nelse:\n" + b + "\n"
            stack.append(expr)
        elif ch in ["_c_"]:            
            stack.append(str(constants[c_counter]))
            c_counter = c_counter + 1
        else:
            stack.append(ch)

    return stack

# if __name__ == '__main__':
#     infix = postfix_to_infix_pyton("cost income - cost return / + cost income - return income / + + cost return cost + - cost return * cost / + * income income return / ^ return return return * income cost * * * * /")
#     print("PG output: " + infix)
#     # print("Simplify: " + str(simplify(infix)))
```
