```python
import pandas as pd
import numpy as np
```

```python
df_features = pd.read_csv("labels.txt", sep = " ", header = None)
for index, row in df_features.iterrows():
    print(row[1] + " | ", end = "")
df_features
```

```python
df_activity = pd.read_csv("activity.txt", sep = " ", header = None)
df_activity.columns = [row[1] for index, row in df_features.iterrows()]
df_activity
```

```python
for c in df_activity.columns[1:-1]:
    df_activity[c] = (df_activity[c] - df_activity[c].mean())/df_activity[c].std()

df_activity
```

```python
df_activity = df_activity.dropna(axis = 1)
```

```python
df_activity.IT_PSD_kurtosis.describe()
```

```python
# 1-neutral, 2-emotional, 3-mental and 4-physical.

df_activity['neutral'] = [1 for i in range(len(df_activity))]
df_activity['emotional'] = [2 for i in range(len(df_activity))]
df_activity['mental'] = [3 for i in range(len(df_activity))]
df_activity['physical'] = [4 for i in range(len(df_activity))]

# df_activity = df_activity[]

# df_activity.columns = 
```

```python
df_activity['physical'], df_activity['activity'] = df_activity['activity'], df_activity['physical']
```

```python
def f(row):
    if(((-0.2>0.2)or(0.2<row['ECG_HR_min_div_mean']))):
        if((-1<0.4)):
            return row['activity']
        else:
            return row['neutral']
    else:
        if((row['IT_p_Total_median']>0)):
            return row['neutral']
        else:
            return row['emotional']
```
