### O ruído é calcular o ri, desvio padrão dos ri e calcular quanto saiu de ruído pra cada amostra

Alterar parâmetros e gramática. Deixar gerar números reais por concatenação

```python
import math

def add_num_noise(column):
    minimum = math.inf
    tries = 1
    best_ei = None

    while tries > 0:
        # F(x)
        f = column

        # median_xF(x)
        mfx = np.median(f)
        
        # E_x|F(x) - median_xF(x)|
        norm = np.mean([fx -  mfx for fx in f])

        # N ratios
        ri = []
        for x in f:
            r = math.inf

            while r == math.inf:
                u = np.random.normal(0, 1)
                v = np.random.uniform(0, 1)
                r = u/v
            
            ri.append(u/v)
        
        # Standard Deviation std_r of the sample
        dri = np.std(ri)

        # Error (noise)
        ei = (norm/dri) * np.sqrt(np.pi/2)
        ei = [r * ei for r in ri]

        # print(ei)
        # print(f)
        if minimum > np.mean(np.abs(ei/f)):
            minimum = np.mean(np.abs(ei/f))
            best_ei = ei
        
        tries = tries - 1
    
    ei = best_ei
    f_new = f + ei
                    
    # print('{:>12}  {:>12}  {:>12}  {:>12}  {:>12} '.format("min(f)", "mean(f)", "median(f)", "std(f)", "max(f)"))
    # print('{:>12.5f}  {:>12.5f}  {:>12.5f}  {:>12.5f}  {:>12.5f}'.format(np.min(f), np.mean(f), np.median(f), np.std(f), np.max(f)))
    
    # print('{:>12}  {:>12}  {:>12}  {:>12}  {:>12}  {:>12}  {:>12}  {:>12}'.format("min(e)", "mean(e)", "median(e)", "std(e)", "max(e)" , "% = mean(e/F(x))", "median(%)", "std(%)"))
    # print('{:>12.5f}  {:>12.5f}  {:>12.5f}  {:>12.5f}  {:>12.5f}  {:>12.5f}  {:>12.5f}  {:>12.5f}'.format(np.min(ei), np.mean(ei), np.median(ei), np.std(ei), np.max(ei), np.mean(np.abs(ei/f)), np.median(np.abs(ei/f)), np.std(np.abs(ei/f))))
    
    return f_new
```

```python
def print_function(f, vars, name):
    uniformly_points = np.random.uniform(low=inferior, high=superior, size=(60000, vars))
    data = []

    for point in uniformly_points:
        point = list(point)
        point.append(f(point))
        data.append(point)

    data = pd.DataFrame(data = data)
    data = data.astype(np.float64)
    data = data.dropna()
    data = data[data[vars - 1] != math.inf]
    data = data[data[vars - 1] != -math.inf]

    test = data.sample(n = 50000)
    data = data[~data.index.isin(test.index)]

    fig, axs = plt.subplots(2, len(amounts), figsize=(16, 10), sharey='row')
    fig.suptitle(name)
    plot = 0

    for amount in amounts:
        # print(vars, name, amount)

        sample = data.sample(n = amount)
        sample.index = ([i for i in range(amount)])
        a = sample.iloc[:, -1].copy()
        axs[0, plot].scatter(x = [i for i in range(len(sample))], y = a, marker='o', label = "f")
        
        ### Without Noise
        # original = pd.concat([pd.DataFrame(data = [[str(vars), "1", "0", str(amount + 50000)], ["x" + str(i + 1) for i in range(vars)] + ["f"]]), sample, test]).reset_index(drop = True)
        # original.to_csv("fs/" + name + "/" + str(amount) + "/original/data", sep = " ", header = False, index = False)
        with open("fs/" + name + "/" + str(amount) + "/original/data", "w") as file:
            file.write(str(vars))
            file.write(" ")
            file.write(str(1))
            file.write(" ")
            file.write(str(0))
            file.write(" ")
            file.write(str(amount + 50000))
            file.write("\n")

            for x in ["x" + str(i + 1) for i in range(vars)]:
                file.write(x)
                file.write(" ")

            file.write("f")
            file.write("\n")

            for index, row in sample.iterrows():
                for i, element in enumerate(row):
                    file.write(str(element))
                    
                    if i < len(row) - 1:
                        file.write(" ")
                    elif index < len(sample) - 1:
                        file.write("\n")
            
            for index, row in test.iterrows():
                for i, element in enumerate(row):
                    file.write(str(element))
                    
                    if i < len(row) - 1:
                        file.write(" ")
                    elif index < len(test) - 1:
                        file.write("\n")
            file.close()

        ### With Noise
        sample.iloc[:, -1] = add_num_noise(sample.iloc[:, -1])
        b = sample.iloc[:, -1].copy()
        axs[0, plot].scatter(x = [i for i in range(len(sample))], y = b, marker='x', label = "f + e")

        # noised = pd.concat([pd.DataFrame(data = [[str(vars), "1", "0", str(amount + 50000)], ["x" + str(i + 1) for i in range(vars)] + ["f"]]), sample, test]).reset_index(drop = True)
        # noised.to_csv("fs/" + name + "/" + str(amount) + "/noise/data", sep = " ", header = False, index = False)
        with open("fs/" + name + "/" + str(amount) + "/noise/data", "w") as file:
            file.write(str(vars))
            file.write(" ")
            file.write(str(1))
            file.write(" ")
            file.write(str(0))
            file.write(" ")
            file.write(str(amount + 50000))
            file.write("\n")

            for x in ["x" + str(i + 1) for i in range(vars)]:
                file.write(x)
                file.write(" ")

            file.write("f")
            file.write("\n")

            for index, row in sample.iterrows():
                for i, element in enumerate(row):
                    file.write(str(element))
                    
                    if i < len(row) - 1:
                        file.write(" ")
                    else:
                        file.write("\n")

            for index, row in test.iterrows():
                for i, element in enumerate(row):
                    file.write(str(element))
                    
                    if i < len(row) - 1:
                        file.write(" ")
                    else:
                        file.write("\n")

            file.close()

        # print('{:>12}  {:>12}  {:>12}'.format("median(a-b)", "std(a-b)", "rmse(a, b)"))
        # print('{:>12.5f}  {:>12.5f}  {:>12.5f}'.format(np.median(a-b), np.std(a-b), np.sum(np.sqrt(np.power(a - b, 2)))))
        
        axs[0, plot].set_title(str(amount))
        axs[0, plot].set_ylabel(name)
        axs[0, plot].legend()

        axs[1, plot].set_title("MSE = " + '{:>.5f}'.format(np.square(a - b).mean()))
        axs[1, plot].plot([i for i in range(len(sample))], np.square(a - b), label = "(a - b)²")
        axs[1, plot].plot([i for i in range(len(sample))], [np.square(a - b).mean() for i in range(len(sample))], label = "mse")
        axs[1, plot].legend()
        axs[1, plot].set_ylabel('SE')

        plot = plot + 1

    plt.savefig("fs/" + name + "/original_vs_noise.svg")
    plt.show()
```
